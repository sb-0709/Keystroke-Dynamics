# Formatting of Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data():
    df = pd.read_csv('C:/Users/Sanskriti/Downloads/DSL-StrongPasswordData.csv')
    
    H_columns  = [col for col in df.columns if col.startswith('H')]
    DD_columns = [col for col in df.columns if col.startswith('DD')]
    UD_columns = [col for col in df.columns if col.startswith('UD')]

    data = {}
    data['total'] = df.drop(columns=['subject', 'sessionIndex', 'rep'])
    data['H']     = df[H_columns]
    data['DD']    = df[DD_columns]
    data['UD']    = df[UD_columns]
    
    return data, df['subject'].values
```

# Visualisation of Dataset

### 1. Keydown- Keydown

```python
data, y = load_data()

# One hot encoding
Y = pd.get_dummies(y).values
n_classes = Y.shape[1]

df_plot = data['DD'].copy()
df_plot['subject'] = y
df_plot = df_plot.groupby('subject').mean()
df_plot.iloc[:8].T.plot(figsize=(14, 7), title='Average Keydown-Keydown Duration')
plt.xlabel('Key Index')
plt.ylabel('Duration');
```

### 2. Pressure Duration (Hold)

```python
df_plot = data['H'].copy()
df_plot['subject'] = y
df_plot = df_plot.groupby('subject').mean()
df_plot.iloc[:8].T.plot(figsize=(14, 7), title='Average Pressure Duration')
plt.xlabel('Key Index')
plt.ylabel('Duration');
```

### 3. Keyup- Keydown

```python
df_plot = data['UD'].copy()
df_plot['subject'] = y
df_plot = df_plot.groupby('subject').mean()
df_plot.iloc[:8].T.plot(figsize=(14, 7), title='Average Keyup-Keydown Duration')
plt.xlabel('Key Index')
plt.ylabel('Duration');
```
